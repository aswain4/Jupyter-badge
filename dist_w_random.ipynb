Use a random number generator to make histograms of popular probability distributions (will do binomial, uniform, and normal). Use the numpy and matplotlib packages.

```python
import numpy as np
from matplotlib import pyplot as plt
```

Normal distribution

```python
mean, std = 10, 0.1
norm = np.random.normal(mean, std, 100)
```

```python
plt.hist(norm)
plt.show()
```

Binomial distribution

```python
trials, prob = 5, 0.5
binom = np.random.binomial(trials, prob, 100)
```

```python
plt.hist(binom)
plt.show()
```

Uniform distribution

```python
uni = np.random.uniform(-1, 0, 1000)
plt.hist(uni)
plt.show()
```
