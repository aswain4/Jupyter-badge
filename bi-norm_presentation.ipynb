### Visual Demonstration that binomial and normal distribution converge in large N limit

##### Also known as the Central Limit Theorem

```python
import numpy as np
from matplotlib import pyplot as plt
```

### First, we will see distributions when N is relatively small

```python
norm1 = np.random.normal(10, 1, 20) #args are mean, std, trials
binom1 = np.random.binomial(20, 0.5, 20) # args are trials, probability, number of tests
# binomial trials = number for 1 test to be complete
plt.subplot(1, 2, 1) #row 1, col 2, index 1
plt.hist(norm1)
plt.title('Normal Distribution-small N')

plt.subplot(1, 2, 2)
plt.hist(binom1)
plt.title('Binomial Distribution-small N')
plt.show()
```

As seen in the above graphs, while the peaks are in relatively similar points along the graph, the graphs are very different.

### Now we will look at moderately sized N

```python
norm2 = np.random.normal(10, 1, 250)
binom2 = np.random.binomial(20, 0.5, 250)
plt.subplot(1, 2, 1) #row 1, col 2, index 1
plt.hist(norm2)
plt.title('Normal Distribution-moderate N')

plt.subplot(1, 2, 2)
plt.hist(binom2)
plt.title('Binomial Distribution-moderate N')
plt.show()
```

As seen in the above graphs, the graphs are starting to look more like one another, but shape is still slightly off.

### Lastly, we will look at the distributions with large N

```python
norm3 = np.random.normal(10, 1, 20000)
binom3 = np.random.binomial(20, 0.5, 20000)
plt.subplot(1, 2, 1) #row 1, col 2, index 1
plt.hist(norm3)
plt.title('Normal Distribution-large N')

plt.subplot(1, 2, 2)
plt.hist(binom3)
plt.title('Binomial Distribution-large N')
plt.show()
```

These graphs are very similar and are almost mirror images of one another.
